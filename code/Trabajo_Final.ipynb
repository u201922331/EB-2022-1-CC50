# **Preprocesado**

Primero verificamos que los IDs de los videos del dataframe coincidan con los del JSON

```python
def checkNotAvailableIDs(j, d):
    jAvailableIDs = []
    for i in j['items']:
        jAvailableIDs.append(int(i['id']))

    notAvailableIDs = []
    for dfIDs in d['category_id']:
        if dfIDs not in jAvailableIDs and dfIDs not in notAvailableIDs:
            notAvailableIDs.append(dfIDs)

    return notAvailableIDs

dataNonExistentIDs = checkNotAvailableIDs(jData, data)
dataNonExistentIDs
```

Usando la función de la celda anterior podemos ver que existen videos que tienen un id de categoria que no existen. En este caso es el ID 29

```python
data[data['category_id'].isin(dataNonExistentIDs)]
```

Por lo que procederemos a extraer dichas categorías de la data disponible para **Estados Unidos** y las agregamos al json de **Francia**

```python
for elem in jDataUS['items']:
    if int(elem['id']) in dataNonExistentIDs and elem not in jData['items']:
        jData['items'].append(elem)

for elem in jData['items']:
    if int(elem['id']) in dataNonExistentIDs:
        print(elem['id'], '->', elem['snippet']['title'])
```

Ahora agregaremos una columna con los títulos de los IDs

```python
categoriesTitles = {}
for item in jData['items']:
    categoriesTitles[int(item['id'])] = item['snippet']['title']

data['category_title'] = [categoriesTitles[id] for id in data['category_id']]
data
```

Adicionalmente, si revisamos la data guardada en el JSON, notaremos que hay algunos ítems con un parámetro **assignable** falso, por lo que también tendremos que descartarlos

```python
def checkNotAssignables(j):
    result = []
    for elem in j['items']:
        if not elem['snippet']['assignable']:
            result.append(int(elem['id']))
    
    return result

notAssignables = checkNotAssignables(jData)
notAssignables
```

```python
data = data[~data['category_id'].isin(notAssignables)]
data
```

Ahora procedemos a modificar la columna **trending_date**, ya que esta, al representar una fecha, debe de estar representada en su respectivo formato.

```python
data['trending_date'] = pd.to_datetime(data['trending_date'], format='%y.%d.%m')
data
```

Hacemos lo mismo para la columna **publish_time**

```python
data['publish_time'] = pd.to_datetime(data['publish_time'], format='%Y-%m-%d').dt.date
data
```

Ahora verificamos si existen datos faltantes

```python
data.isna().sum()
```

Tal como se puede apreciar, todos las columnas, salvo **description**, están completas. Los vacíos en esta columna no suelen ser inusuales, dado que existe la posibilidad de que un video no tenga descripción. Sin embargo, es preferible llenar dichos vacíos con un **""**.

```python
data['description'].fillna("", inplace=True)
data
```

Viendo la descripción de datos de **video_id**, podemos apreciar que hay un total de **10,143** videos repetidos en nuestro dataframe, y esto se debe gracias a la variable **trending_date**, por lo que nos quedaremos únicamente con el último registro de cada uno de los videos repetidos, que son aquellos con la información más actualizada de los mismos

Usaremos una función llamada **drop_duplicates**, se usará como criterio de comparación la columna **video_id** y se ingresará como parámetro **keep='last'** para conservar el último de cada uno.

```python
data.drop_duplicates(subset=['video_id'], keep='last', inplace=True)
data
```

# Adicionales

* ¿Es factible predecir el número de "Vistas" o "Me gusta" o "No me gusta"?

```python
dataViewsPred=data.copy()[['category_id','views']]
dataViewsPred.sort_values('category_id',ascending=True,inplace=True)

dataViewsPred=dataViewsPred.groupby('category_id')
dvp=pd.DataFrame()
dataMin=dataViewsPred['views'].min()

arrMinViews=dataMin

dvp['menos']=arrMinViews

dataMax=dataViewsPred['views'].max()
arrMaxViews=dataMax
dvp['max']=arrMaxViews

dataProm=dataViewsPred['views'].mean()
arrPromViews=dataProm
dvp['prom']=arrPromViews
dvp.reset_index(inplace=True)
print(dvp)
```

```python
df=pd.DataFrame(dvp,columns=['category_id','menos','max','prom'])
df.plot(x='category_id', y=['menos','max','prom'],kind='bar')
plt.show()
```

Se puede apreciar que existe una gran diferencia entre los valores minimos, promedio y maximos de las visitas por categoria, por lo que intentar simular cuantas visitas se podrian obtener seria algo arriesgado.
